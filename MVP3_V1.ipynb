A Olist é uma das principais plataformas de comércio eletrônico do Brasil, e esse MVP visa contribuir significativamente para aprimorar a experiência tanto para os clientes quanto para os vendedores na plataforma.

###Problema a ser Abordado:

Eficiência Logística:
Um dos problemas centrais enfrentados diz respeito à eficiência logística no processamento e entrega de pedidos. O tempo de entrega de produtos adquiridos online tem um impacto direto na satisfação do cliente e na reputação da empresa. Pretende compreender melhor os principais fatores que afetam o tempo de entrega e identificar áreas geográficas que enfrentam os maiores desafios nesse aspecto. Avaliar se existe uma correlação entre a distância entre o vendedor e o cliente e o tempo de entrega. Resolver esses desafios contribuirá para entregas mais rápidas e confiáveis, resultando em clientes mais satisfeitos.

Satisfação do Cliente:

Isso nos permitirá tomar medidas específicas para melhorar a experiência do cliente, identificando áreas que necessitam de atenção.

Desempenho do Produto e dos Vendedores:
Outro aspecto importante é o desempenho dos produtos e dos vendedores na plataforma da Olist. Queremos identificar os produtos mais vendidos e entender quais categorias de produtos enfrentam as maiores taxas de devolução. Além disso, planejamos analisar o desempenho dos vendedores, identificando os que têm as maiores taxas de vendas, bem como os que recebem mais reclamações de clientes. Isso nos ajudará a orientar os vendedores para um melhor desempenho e, assim, aumentar a satisfação do cliente.

Em resumo, o nosso MVP em engenharia de dados tem como objetivo principal ajudar a entender como resolver questões relacionadas à eficiência logística e à satisfação do cliente no comércio eletrônico da Olist. ## Como foi feita a Engenharia de Dados para Análise de E-commerce da Olist

Em um esforço para explorar e analisar dados de e-commerce de maneira mais eficaz, realizei uma engenharia de dados detalhada, combinando várias fontes de dados relevantes em um único conjunto de dados coeso. A base de dados que utilizei é proveniente da Brazilian E-Commerce Public Dataset by Olist, que contém uma ampla variedade de informações sobre pedidos, clientes, produtos e vendedores.

Minha primeira etapa foi a seleção das informações mais relevantes para o meu projeto, o que resultou em um conjunto de dados mais focado e gerenciável.

##A seguir, descrevo as etapas de engenharia de dados que realizei:

###Seleção de Colunas Essenciais:

Inicialmente, selecionei colunas essenciais de diferentes conjuntos de dados para serem integradas ao meu DataFrame principal. Essas colunas incluem informações sobre pedidos, como 'order_id', 'order_status', 'Review_score', 'customer_city', 'product_category_name', 'days_between', 'days_behind' bem como dados de avaliações de pedidos, incluindo 'review_score'.

###Integração de Dados de Clientes:

 Em seguida, integrei informações de clientes, incluindo 'customer_city', 'customer_state' e 'customer_id'. Esses dados são cruciais para entender o perfil dos clientes que realizaram os pedidos.

###Incorporação de Dados de Itens do Pedido:

Adicionei informações relacionadas aos itens de cada pedido, como 'product_id', 'seller_id', 'price' e 'freight_value'. Esses detalhes são importantes para avaliar aspectos de preços e frete em relação aos pedidos.

###Inclusão de Dados de Vendedores:

Acrescentei dados de vendedores, como 'seller_city' e 'seller_state', que são valiosos para entender a localização geográfica dos vendedores.

###Complementação com Categorias de Produtos:

 Finalmente, integrei dados de categorias de produtos ('product_category_name') para categorizar os produtos associados a cada pedido.

###Estimativa de Entrega e Conformidade:

Também incluí informações sobre a estimativa de entrega e se essa estimativa foi atingida ou não. Isso é fundamental para avaliar a eficiência logística da Olist.


A fusão cuidadosa dessas informações resultou no DataFrame merged_df_2, que é uma representação consolidada e pronta para análise de várias facetas do e-commerce da Olist. Com essa abordagem de engenharia de dados, agora tenho um conjunto de dados unificado que me permitirá explorar e responder a uma série de perguntas importantes relacionadas à eficiência logística, satisfação do cliente, desempenho de produtos e vendedores, entre outros aspectos cruciais do negócio.

Esse processo de engenharia de dados é fundamental para preparar os dados de forma adequada para análises posteriores em meu projeto de ciência de dados. Com um conjunto de dados bem estruturado e abrangente, estou pronto para prosseguir com análises detalhadas e obter insights valiosos que contribuirão para melhorar a eficácia e o desempenho da plataforma de e-commerce da Olist.

```python
def load_to_gcp(storage_client, bucket_name, destination_blob_name, filename):

    filename.to_csv('temp.csv', index=False)

    bucket = storage_client.bucket(bucket_name)
    blob = bucket.blob(destination_blob_name)

    # Carregar o arquivo CSV para o GCP
    blob.upload_from_filename('temp.csv')

    print(f"Arquivo {filename} carregado para {destination_blob_name}.")

# In your main function
storage_client = storage.Client(credentials=credentials)
```

Realisando o downloads das bases GTP

```python
bucket = client.get_bucket(bucket_name)
customer = bucket.get_blob(customers_name)
order_items = bucket.get_blob (order_name)
order_reviews = bucket.get_blob (reviews_name)
products_name = bucket.get_blob (products_name)
sellers_name = bucket.get_blob (sellers_name)
orders_name = bucket.get_blob (orders_name)
```

```python
data_customers = customer.download_as_string()
data_order_items = order_items.download_as_string()
data_reviews_name = order_reviews.download_as_string()
data_products_name = products_name.download_as_string()
data_sellers_name = sellers_name.download_as_string()
data_orders_name = orders_name.download_as_string()
```

```python
df_data_customers = pd.read_csv(io.StringIO(data_customers.decode('utf-8')))#
df_data_order_items  = pd.read_csv(io.StringIO(data_order_items.decode('utf-8')))#
df_data_order_reviews = pd.read_csv(io.StringIO(data_reviews_name.decode('utf-8')))#
df_data_product_name = pd.read_csv(io.StringIO(data_products_name.decode('utf-8')))#
df_sellers_name = pd.read_csv(io.StringIO(data_sellers_name.decode('utf-8')))#
df_data_orders_name = pd.read_csv(io.StringIO(data_orders_name.decode('utf-8')))#
```

```python
# Lista de todos os DataFrames e as colunas para mesclar
dfs = [
    (df_data_orders_name[['order_id', 'customer_id', 'order_status', 'order_purchase_timestamp','order_delivered_customer_date', 'order_estimated_delivery_date']], 'order_id'),
    (df_data_order_reviews[['review_score','order_id']], 'order_id'),
    (df_data_customers[['customer_city', 'customer_state', 'customer_id']], 'customer_id'),
    (df_data_order_items[['product_id', 'seller_id', 'price','freight_value','order_id']], 'order_id'),
    (df_sellers_name [['seller_city', 'seller_id', 'seller_state']], 'seller_id'),
    (df_data_product_name[['product_category_name', 'product_id']], 'product_id')
]

# Realize todas as operações de mesclagem de uma vez
merged_df = dfs[0][0]
for df, merge_col in dfs[1:]:
    merged_df = pd.merge(merged_df, df, on=merge_col, how='left')

# Transforme as colunas de data
for col in ['order_delivered_customer_date', 'order_purchase_timestamp', 'order_estimated_delivery_date']:
    merged_df[col] = pd.to_datetime(merged_df[col])

# Crie novas colunas
merged_df['days_between'] = (merged_df['order_delivered_customer_date'] - merged_df['order_purchase_timestamp']).dt.days
merged_df['days_behind'] = (merged_df['order_delivered_customer_date'] - merged_df['order_estimated_delivery_date']).dt.days

# Descarte as colunas desnecessárias
cols_to_drop = ['customer_id', 'order_purchase_timestamp', 'order_delivered_customer_date', 'order_estimated_delivery_date', 'product_id', 'seller_id']
merged_df.drop(cols_to_drop, axis=1, inplace=True)

merged_df.head(3)
```

2.Existe uma correlação entre o valor do frete e o tempo de entrega?

```python
correlacao = merged_df_2[['days_between', 'freight_value']].corr()
print("A correlação entre as colunas 'days_between' e 'frete' é:\n", correlacao)
```

Nem sempre existe correlação entre o valor do frete e tempo de entrega. Visto que o valor do frete não é só em relação à distância, mas tambem ao peso e volume dos produtos.


3.Quais áreas geográficas (cidades ou estados) têm os vendedores com os maiores atrasos na entrega?

```python
total_linhas = merged_df_2['seller_state'].value_counts()
df_filtrado = merged_df_2[merged_df_2['days_behind'] > 0]
linhas_positivas = df_filtrado['seller_state'].value_counts()
percentual = (linhas_positivas / total_linhas) * 100
resultado = pd.DataFrame({
    'Total de Linhas': total_linhas,
    'Total de Linhas Positivas': linhas_positivas,
    'Percentual (%)': percentual
})
resultado = resultado.sort_values('Percentual (%)', ascending=False)
print(resultado)
```

Amazonas, Maranhão e Rio grande do Norte, tem o maior percentual de vendedores com entregas em atraso, porém os mais significativos são São Paulo, Rio de Janeiro e Paraná, por terem o maior numero de vendas.

4.Em quais cidades moram os compradores que mais são afetados pelos atrasos das entregas.

```python
total_linhas = merged_df_2['customer_state'].value_counts()
df_filtrado = merged_df_2[merged_df_2['days_behind'] > 0]
linhas_positivas = df_filtrado['customer_state'].value_counts()
percentual = (linhas_positivas / total_linhas) * 100
resultado = pd.DataFrame({
    'Total de Linhas': total_linhas,
    'Total de Linhas Positivas': linhas_positivas,
    'Percentual (%)': percentual
})
resultado = resultado.sort_values('Total de Linhas Positivas', ascending=False)
print(resultado)
```

O gráfico apresenta as vendas e as médias de review score de top 10 das categorias de produtos mais vendidos. As barras azuis representam as vendas totais para cada categoria de produto, enquanto a linha vermelha representa a média do review score para cada categoria. A categoria com as maiores vendas é “Cama mesa e banho” e ao mesmo tempo é a categoria com a menor review score das categorias mais vendidas. A categoria com maior review score, resultou ser "beleza e saude" sendo a segunda categoria mais vendida a segunda.

10.Existe uma correlação entre a categoria do produto (como "esporte e lazer", "utilidades domésticas" etc.) e o tempo médio de entrega (days_between)? Algumas categorias de produtos têm tempos de entrega significativamente mais longos ou curtos do que outras?

11.Como a distância entre a cidade do cliente (customer_state) e a cidade do vendedor (seller_state) afeta a eficiência logística?
 Existem categorias de produtos que são frequentemente enviadas de estados ou cidades distantes, resultando em frete mais alto ou maior tempo de entrega?



```python
matrix = merged_df_2.pivot_table(index='customer_state', columns='seller_state', values='days_between', aggfunc='mean')

plt.figure(figsize=(10, 6))
sns.heatmap(matrix, annot=True, cmap="YlGnBu")

plt.title('Mapa de calor das médias de day_between')
plt.show()
```

Existem alguns pares de estados com tempos de entrega particularmente altos. Por exemplo, pedidos de clientes em AL (Alagoas) para vendedores em AM (Amazonas) mostram um valor de 90.
Em contraste, outros pares interestaduais têm tempos de entrega mais curtos. Por exemplo, pedidos de SP (São Paulo) para vendedores em RJ (Rio de Janeiro) têm uma média de 11.

```python
matrix = merged_df_2.pivot_table(index='customer_state', columns='seller_state', values='freight_value', aggfunc='mean')

plt.figure(figsize=(10, 6))
sns.heatmap(matrix, annot=True, cmap="YlGnBu")

plt.title('Mapa de calor das médias de frete')
plt.show()
```

Quanto maior a distância entre cliente e vendedor, geralmente maior é o frete. O mapa mostra frete mais baixo entre estados próximos e mais alto entre estados distantes.
